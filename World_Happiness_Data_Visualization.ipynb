```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('World_Happiness_2015_2017_.csv')
data.head()
```

```python
data.dtypes
```

```python
# checking for null values
data.isnull().sum()
```

```python
# checking correlation between attributes
g = sns.pairplot(data)
g.fig.suptitle('FacetGrid Plot', fontsize= 20)
g.fig.subplots_adjust(top= 0.9);
```

```python
# creating a list of Happiness Score and Family
fam_score = ['Happiness Score', 'Family']

# Dataframe that contains the two attributes only
fam_score_data = data[fam_score]

# check for correlation
fam_score_data.corr()
```

```python
sns.regplot(fam_score_data, x='Family', y='Happiness Score').set_title('Correlation graph for Happiness Score vs Family')
```

```python
# creating a list of Happiness Score and Health (Life Expectancy)
life_score =['Happiness Score', 'Health (Life Expectancy)']

life_score_data = data[life_score]
life_score_data.corr()
```

```python
sns.regplot(life_score_data, x='Health (Life Expectancy)', y='Happiness Score').set_title('Correlation graph for Happiness Score vs Health (Life Expectancy)')
```

```python
# creating a list of Happiness Score and freedom
free_score= ['Happiness Score', 'Freedom']
free_score_data= data[free_score]

free_score_data.corr()
```

```python
sns.regplot(free_score_data, x='Freedom', y='Happiness Score').set_title('Correlation graph for Happiness Score vs Freedom')
```

```python
# creating a list of Happiness Score and Trust (Government corruption)
trust_score = ['Happiness Score', 'Trust (Government Corruption)']

trust_score_data = data[trust_score]
trust_score_data.corr()
```

```python
sns.regplot(trust_score_data, x='Trust (Government Corruption)', y= 'Happiness Score').set_title('Correlation graph for Happiness Score vs Trust')
```

```python
sns.regplot(gene_score_data, x='Generosity', y='Happiness Score').set_title('Correlation graph for Happiness Score vs Generosity')
```
